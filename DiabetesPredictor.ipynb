```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
```

```python
x=scaler.transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.2,random_state=2)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

```python
x_train_predicted=classifier.predict(x_train)
training_data_accuracy=accuracy_score(x_train_predicted,y_train)
training_data_accuracy
```

```python
x_test_predicted=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_predicted,y_test)
test_data_accuracy
```

```python
input_data=(4,110,92,0,0,37.6,0.191,30)
input_data_=np.asarray(input_data)
input_data_reshaped=input_data_.reshape(1,-1)
std_data=scaler.transform(input_data_reshaped)
prediction=classifier.predict(std_data)
if(prediction[0]==0):
    print("The person is not diabetic")
else:
    print("The person is diabetic")
```
